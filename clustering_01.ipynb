```python
def DFS(T,i):
    '''
    Deep-First Search algortihm
    
    Input:
            T is the input graph
            i is the node that algorithm finds all reachable vortecies from it in the T
    Outputs:
            T_explored is a dictionary represnting each node with values True/False as explored/unexplored
    '''
    
    # initializing Global variables
    global T_explored

    T_explored[str(i)] = True # make i explored
    for j in T[str(i)]:
        if T_explored[str(j)] == False:
            DFS(T,j)
            
    return T_explored
```

```python
# reading the input file as a list
with open('clustering1.txt') as f:
    reader = csv.reader(f, delimiter=" ")
    d = list(reader)

k = 4 # number of clusters
n = int(d.pop(0)[0]) # number of points, pop first line of file (which is the number of points)

       
# sort d based on distance
d.sort(key=lambda x: int(x[2]))

# is adjancency list of the spanning tree after clustering (it is not fully connected)
T = {} 
for i in range(1,n+1):
    T[str(i)] = []
    
num_clusters = n # current number of clusters, initially each node is a cluster = n

# iteration over all edges
for i in d:
    # check if there is a path between i[0] and i[1]
    T_explored = {x: False for x in T} 
    T_explored = DFS(T,int(i[0])) # DFS function works on graphs given as adjancency list
    if T_explored[str(i[1])] == False: # check if i[1] is reachable from i[0]
        #  if it is not reachable from i[0], then add vertices of edge to T
        T[str(i[0])].append(int(i[1]))
        T[str(i[1])].append(int(i[0]))
        num_clusters -= 1 # update number of clusters after merging
        if num_clusters == k: # clustering is finished
            T_cluster = copy.deepcopy(T) # save the spanning tree as T_cluster
        
        # if we reached to the desired number of clusters, continue one more time to find Maximum_Spacing 
        if num_clusters == k-1:
            Max_Spacing = int(i[2]) # i is the edge that would be added if continue one more time, i[2] is its distance (cost)
            break
            
print('Number of clusters (input): {} \n'.format(k))
print('The maximum spacing: {} \n'.format(Max_Spacing))
print('The spaning tree: \n{}'.format(T_cluster))

# to find the clusters, we can run SCC on T_cluster
```
